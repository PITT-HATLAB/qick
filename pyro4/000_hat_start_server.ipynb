# Start a pyro server on the RFSoC board

* `ns_host` should be the ip address of the name server, which usually runs on the local windows computer that you work on.\
You can copy that address from the returns of the "startNameserver.bat" program. e.g.\


* `proxy_name` is the identifier that the client will use to find this server.\
 You can choose whatever proxy name you want, as long as you use the same one on the client (the Windows PC, in the yml file).\
  **Make sure each board has a unique proxy_name!!!**


```python
from qick.pyro import start_server
start_server(ns_host="192.168.1.5", ns_port=8888, proxy_name="myqick111_01")
```
